```python
# Example of if-elif-else statements
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")

# You can change the 'score' variable to test different conditions.
```

This `if-elif-else` structure allows you to check multiple conditions sequentially. The first condition that evaluates to `True` will have its corresponding code block executed, and the rest of the `elif` and `else` blocks will be skipped.

```python
# if-elif-else statements
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# Energy Price Calculation

units = float(input("Enter the number of units consumed: "))

print("Customer Types:")
print("1. Residential")
print("2. Commercial")
print("3. Industrial")

choice = int(input("Enter your choice (1-3): "))

if choice == 1:
    price = units * 2
    print("Customer Type: Residential")
    print("Energy Price = ₹", price)
else:
    if choice == 2:
        price = units * 5
        print("Customer Type: Commercial")
        print("Energy Price = ₹", price)
    else:
        if choice == 3:
            price = units * 10
            print("Customer Type: Industrial")
            print("Energy Price = ₹", price)
        else:
            print("Invalid customer type.")
```
